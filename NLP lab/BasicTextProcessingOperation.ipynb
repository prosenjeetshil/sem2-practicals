```python
def punctuation(data):
  text=data
  text=text.replace("fucking","failing")
  text=text.replace("sucking","falling")
  text=text.replace("sucking","dying")
  return text

s="i love fucking, sucking, sucking"
returned_data = punctuation(s)
print(returned_data)
```
